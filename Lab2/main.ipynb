# Questão 3

```python
lista1 = np.array([11, 13, 13, 13, 13, 13, 9, 13, 13, 13, 9, 13, 13, 13, 13, 13, 13, 13, 13, 13, 9, 13, 11, 13, 9, 11, 11, 13, 11, 11])
lista2 = np.array([11, 13, 13, 6, 11, 11, 9, 13, 13, 11, 13, 11, 13, 13, 6, 11, 11, 11, 13, 13, 6, 13, 13, 11, 13, 6, 13, 9, 13, 13])

def calcular_estatisticas(lista):
    media = np.mean(lista)
    desvio = np.std(lista)
    n = len(lista)
    intervalo = sp.stats.norm.interval(0.95, loc=media, scale=desvio/np.sqrt(n))
    return media, desvio, intervalo

media1, desvio1, intervalo1 = calcular_estatisticas(lista1)
media2, desvio2, intervalo2 = calcular_estatisticas(lista2)

print(f"Lista 1: Média = {media1:.2f}, Desvio Padrão = {desvio1:.2f}, Intervalo de Confiança = {intervalo1[0]:.2f}, {intervalo1[1]:.2f}")
print(f"Lista 2: Média = {media2:.2f}, Desvio Padrão = {desvio2:.2f}, Intervalo de Confiança = {intervalo2[0]:.2f}, {intervalo2[1]:.2f}")
```

# Questão 8

```python
x = np.arange(0, 5.2, 0.2)
y_linear = x
y_quadratic = x**2
y_cubic = x**3

plt.figure(figsize=(10, 6))
plt.plot(x, y_linear, 'ro-', label='Linear (vermelho)')
plt.plot(x, y_quadratic, 'bs-', label='Quadrática (azul)')
plt.plot(x, y_cubic, 'g^-', label='Cúbica (verde)')
plt.title("Gráfico com variação linear, quadrática e cúbica")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

# Questão 12

```python
titanic = sns.load_dataset("titanic")
survival_rate = titanic.groupby(['sex', 'class'], observed=False)['survived'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=survival_rate, x='class', y='survived', hue='sex')
plt.title("Taxa de Sobrevivência por Sexo e Classe - Titanic")
plt.ylabel("Taxa de Sobrevivência")
plt.ylim(0, 1)
plt.show()
```

# Questão 15

```python
iris_data = load_iris()
X = iris_data.data
y = iris_data.target

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)
tsne_df = pd.DataFrame(X_tsne, columns=["TSNE1", "TSNE2"])
tsne_df['species'] = pd.Categorical.from_codes(y, iris_data.target_names)

plt.figure(figsize=(8, 6))
sns.scatterplot(data=tsne_df, x="TSNE1", y="TSNE2", hue="species", palette="Set1")
plt.title("Projeção T-SNE da base Iris")
plt.show()
```
